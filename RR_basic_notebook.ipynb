```python
import pandas as pd

# Create a DataFrame with one column 'power' ranging from 0 to 50
df = pd.DataFrame({'power': range(0, 51)})

# Save the DataFrame to a CSV file
df.to_csv('power_data.csv', index=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('power_data.csv')

# Add white noise (mean=0, std=2 for example)
noise = np.random.normal(loc=0, scale=2, size=len(df))
df['power_noisy'] = df['power'] + noise

# Plot original and noisy data
plt.figure(figsize=(10, 5))
plt.plot(df['power'], label='Original Power')
plt.plot(df['power_noisy'], label='Noisy Power', linestyle='--')
plt.xlabel('Index')
plt.ylabel('Power')
plt.title('Power with White Noise')
plt.legend()
plt.grid(True)
plt.show()
```
